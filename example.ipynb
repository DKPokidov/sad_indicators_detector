```python
district_list = df['properties.name'].tolist()

for index in range(len(df)):
    sir_indicators = pd.DataFrame(df['properties.indicators'][index])
    sir_indicators = sir_indicators.sort_values(by='name_full', ascending=True)
    
    values = sir_indicators['value'].values
    district = district_list[index]
        
    if len(values) < len(sad_table):
        values = np.concatenate([values, [None] * (len(sad_table) - len(values))])
    # elif len(values) > len(sad_table):
    #     values = values[:len(sad_table)]
    sad_table[district] = values
```

```python
sad_table.sort_index(ascending=True)
```

```python

def change_func(value):
    if pd.isna(value):
        return value
    elif value > 0.0:
        return int(1)
    elif value == 0.0:
        return int(0)
    else:
        return None
```

```python
sad_table['Сосновоборский городской округ'] = sad_table['Сосновоборский городской округ'].astype('float64')
```

```python
sad_table.info()
```

```python
table = sad_table.apply(lambda col: col.apply(change_func))
```
